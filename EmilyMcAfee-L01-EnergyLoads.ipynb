```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cleaned up Dataset location
fileName = "https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/EnergyEfficiencyData.csv"
```

```python
# Read in dataset/csv as dataframe
buildings = pd.read_csv(fileName)
# Look at data
buildings.head()
buildings.dtypes
#list(buildings)
```

```python
fig3 = plt.figure(figsize=(6,6)) # Define plot area
ax3 = fig3.gca() # Define axis 
sns.set_style("whitegrid")
sns.violinplot(x = 'Overall Height', y = 'Heating Load', data = buildings, ax = ax3,  split = True)
ax3.set_title('Violin plot of Heating Load by Overall Height')
ax3.set_xlabel('Overall Height')
ax3.set_ylabel('Heating Load')# Set text for y axis

print('The heating load of buildings depends on the overall height. Evidence for this relationship can be seen in the figure immediately below by noting the higher heating load for an overall height of 7.0 compared to 3.5.')
```

```python
fig2 = plt.figure(figsize=(6, 6)) # Define plot area
ax2 = fig2.gca() # Define axis 
buildings.loc[:,['Heating Load', 'Glazing Area']].boxplot(by = 'Glazing Area', ax = ax2)
ax2.set_title('Box plot of Heating Load by Glazing Area')
ax2.set_xlabel('Glazing Area')
ax2.set_ylabel('Heating Load')

print('The heating load of buildings can depend on the glazing area. Evidence for this relation can be seen in the boxplot below by noting the median of each glazing area group (i.e. 0.0, 0.1, .25, .4) as proportionally increasing as the heating load increases.')






```

```python
fig = plt.figure(figsize=(6,6)) # Define plot area
ax = fig.gca() # Define axis 
sns.set_style("whitegrid")
sns.kdeplot(buildings.loc[:, 'Heating Load'], ax = ax)
ax.set_title('KDE plot of Heating and Cooling Load') # Give the plot a main title
ax.set_xlabel('Load') # Set text for the x axis
ax.set_ylabel('Density')# Set text for y axis

sns.kdeplot(buildings.loc[:, 'Cooling Load'], ax = ax)

print('Interestingly, the heating load is also related to the cooling load. Both load measures are very similar - although not exactly the same - as noted in the KDE plot below. By knowing the heating load of the building, cooling can be reliably estimated. This also works vice versa (i.e. if cooling is known the heating can be reliably estimated).')






```

## Key Points:

**2)** The heating load of buildings can depend on the glazing area. Evidence for this relation can be seen in the boxplot above by noting the median of each glazing area group (i.e. 0.0, 0.1, .25, .4) as proportionally increasing as the heating load increases.
